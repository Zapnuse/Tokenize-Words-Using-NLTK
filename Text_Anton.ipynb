```python
import nltk
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Text Mining SI 20C/new_data_Fintech(2).csv')
data.to_csv('new_file_before_preprocessing.csv', index = False)
data['article'].head()
```

```python
#clean empty row/Nan
data.dropna(subset = ["article"], inplace=True)

data.to_csv('new_file_without_empty_rows.csv', index=False)
```

```python
import nltk
nltk.download('punkt')
```

```python
#menggabungkan semua data dan dipisahkan menggunakan space
datanew = data.article.str.cat(sep='')
print (datanew)
```

```python

#Tokenize words
words = nltk.tokenize.word_tokenize(datanew)

#count word
word_dist = nltk.FreqDist(words)

#nampilkan 10 kata pertama dengan frequency terbanyak
top_N = 100

#nampilkan kata dalam table menghitung frquency 
print("Word Frequency before remove stopword")
print('All Frequencies, including STOPWORDS:')
print('=' * 60)
rslt = pd.DataFrame(word_dist.most_common(top_N),
                    columns=['word', 'Frequency']).set_index('word')

print(rslt)
print('=' * 60)
```
